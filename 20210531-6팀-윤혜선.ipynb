```python
#최소값 구하기
n1=int(input('숫자1:'))
n2=int(input('숫자2:'))
n3=int(input('숫자3:'))
#elif를 쓰면 안됨!!
#elif를 쓰면 다음것을 확인하지 않고 출력해버리기 때문에
#모든 숫자를 체크해야하는경우 if를 써야한다.
min1=n1
if n2<min1:
    min1=n2
if n3<min1:
    min1=n3
print('가장작은수:', min1)
```

```python
#최소값 구하기 min함수를 사용하여
n1=int(input('숫자1:'))
n2=int(input('숫자2:'))
n3=int(input('숫자3:'))

min_n=min(n1,n2,n3)

print('가장 작은 수:',min_n)
```

```python
#3번 예제 사칙연산 프로그램

num1=int(input('숫자1:'))
num2=int(input('숫자2:'))
sig=input('연산기호(+,-,*,/):')

#if 문을 쓰면 각각의 문장을 전부 체크하게된다.
#elif보다 if 문을 쓰는게 나을수 있다. (여기서는)

if sig=='+':
    print(num1+num2) 
if sig=='-':
    print(num1-num2)
if sig=='*':
    print(num1*num2)
if sig=='/':
    if num2 !=0:
        print(float(num1/num2))
    else:
        print('NONE')
```

```python
#3번 예제 사칙연산 프로그램

num1=int(input('숫자1:'))
num2=int(input('숫자2:'))
sig=input('연산기호(+,-,*,/):')

if sig=='+':
    print(num1+num2) 
elif sig=='-':
    print(num1-num2)
elif sig=='*':
    print(num1*num2)
elif sig=='/':
    if num2 !=0:
        print(float(num1/num2))
    else:
        print('0으로 나눌 수 없습니다')
```
