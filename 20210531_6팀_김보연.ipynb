```python
#편의점 재고 관리 
inventory = {}
#물건이름:product, 개수:number
while True:
    product = input('물건이름:')
    if product == '':
        break
    number = int(input('number: '))
    if number in inventory.keys():
        inventory[product] += number
    else:
        inventory[product] = number
#오름차순 정렬 
for key,value in sorted(inventory.items()):
    print(key, value, end=' ')
```

```python
sweety = {'chocolate', 'jelly', 'caramel'}
salty = {'potatochip', 'honeybutterchip', 'chichenpop', 'caramel'}

# 합집합
print(sweety.union(salty))
print(sweety|salty)

# 교집합
print(sweety.intersection(salty))
print(sweety&salty)

# 차집합(앞의 집합에 따라 결과가 달라진다)
print(sweety.difference(salty))
print(sweety-salty)
print(salty.difference(sweety))
print(salty-sweety)
```

```python
#예제>코로나 자가진단 프로그램 
temp=float(input('체온:'))
check=(input('검사대기여부(y or n):'))
isol=(input('자가격리여부(y or n):'))
if temp >37.5:
    print('발열기준초과:등교불가')
elif check=='y':
    print('검사대기 중 등교불가')
elif isol=='y':
    print('자각격리 중 등교불가')
else:
    print('등교가능')
```

```python
#예제>점수에 따른 등급 출력하기 
socre = int(input('점수:'))
if socre >=90:
    grade = 'A'
elif socre >=80:
    grade = 'B'
elif socre >=70:
    grade = 'C'
elif socre >=60:
    grade = 'D'
else:
    grade = 'F'   
print(f'등급:{grade}')
```

```python
#예제>금액에 따른 할인율 계산하기
price = int(input('물건 구매가를 입력하세요: '))

if price >= 100000:
    per = 0.1
elif price >= 50000:
    per = 0.075
elif price >= 10000:
    per = 0.05
else:
    per = 0   #구간에 없는 부분도 신경쓰자

dc = price*per
print(f'구매가: {price}원')
print(f'할인율: {per*100}%')
print(f'할인금액: {int(dc)}원')   #int를 씌워서 소수점 아래는 버리고 깔끔하게 나오기 위함
print(f'지불금액: {int(price-dc)}원')
```

```python
#예제>최소값 구하기 
n1 = int(input('숫자 1:'))
n2 = int(input('숫자 2:'))
n3 = int(input('숫자 3:'))
min1= n1 
if n2 <min1:
    min1 = n2

if n3 < min1:
    min1 = n3
print('가장 작은 수:', min1)
```

```python
#예제>최소값 구하기 
n1 = int(input('숫자 1:'))
n2 = int(input('숫자 2:'))
n3 = int(input('숫자 3:'))

min_n =min(n1,n2,n3)
print('가장 작은 수:', min_n)
```
