```python
d_li={}
while True:
    name=input('물건이름:')
    if name=='':
        break
    num=int(input('개수:'))
    if name in d_li.keys():
        d_li[name]+=num
    else:
        d_li[name]=num
for key, value in sorted(d_li.items()):
    print(f'{key}: {value}', end=' ')
```

```python
snack1={'빼빼로', '칸초', '허니버터'}
snack2={'허니버터', '포카칩', '새우깡'}

print(snack1.union(snack2))
print(snack1|snack2)
print(snack1.intersection(snack2))
print(snack1&snack2)
print(snack1.difference(snack2))
print(snack1-snack2)
print(snack2.difference(snack1))
print(snack2-snack1)
```
