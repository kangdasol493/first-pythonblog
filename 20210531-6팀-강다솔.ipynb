예제1_편의점 재고 관리

```python
#물건이름:product, 개수:number
dict_li={}

while True:
    product= input('물건이름: ')
    if product == '':
        break
    number = int(input('개수: '))
    if product in dict_li.keys():
        dict_li[product]+=number
    else:
        dict_li[product]=number
print(dict_li)
#오름차순 정렬
for key,value in sorted(dict_li.items()):
    print(key,value,end=' ')
```

집합(set)_요소추출
* 집합자료형은 순서가 없기 때문에 인덱싱으로 값을 얻을 수 없다.
* 집합 자료형에 저장된 값을 인덱싱으로 접근하려면 리스트나 튜플로 변환하여 추출한다.




```python
set_chip={'포테토칩','포카칩','수미칩','포테토칩'}
list_schip=list(set_chip)
print(list_schip)
print(list_schip[2])
```

```python
set_chip={'포테토칩','포카칩','수미칩','포테토칩'}
list_schip=tuple(set_chip)
print(list_schip)
print(list_schip[2])
```

```python
# a와 b의 합집합 a.union(b)         a|b
# a와 b의 교집합 a.intersction(b)   a&b
# a와 b의 차집합 a.difference(b)    a-b
```

```python
sweety={'빼빼로','칸초','허니버터'}
salty={'허니버터','포카칩','새우깡'}

# 합집합
print(sweety.union(salty))
print(sweety|salty)

# 교집합
print(sweety.intersection(salty))
print(sweety&salty)

# 차집합
print(sweety.difference(salty))
print(sweety-salty)
print(salty.difference(sweety))
print(salty-sweety)
```

**오류메세지 해석**
*  NameError : 참조변수 없음(not defined)
*  SyntaxError : 문법이 잘못되었음 (오타,괄호를 제대로 안 닫았을 때, 콜론(:)을 잘못 사용했을 때 등...)
*  TypeError : 타입이 맞지 않음
*  ZeroDivisionError : 0으로 나눌 수 없음
*  IndentationError : 들여쓰기 오류

★ **프로그램의 3가지 기본 제어 구조**_
순차구조, 선택구조, 반복구조

```python
#예제_금액에 따른 할인율 계산하기

price=int(input('물건 구매가를 입력하세요 : '))

if price>=100000:
    dc=10
elif price>=50000:
    dc=7.5
elif price>=10000:
    dc=5
else:
    dc=0

print(f'구매가 : {price}원')
print(f'할인율:{dc}%')
print(f'할인금액:{int(price*dc/100)}원')
print(f'지불금액:{int(price-(price*dc/100))}원')
```

```python
#예제_사칙연산 프로그램

num1=int(input('숫자1 : '))
num2=int(input('숫자2 : '))
oper=input('연산기호(+,-,*,/) : ')

if oper=='+':
    print(num1+num2)
elif oper=='-':
    print(num1-num2)
elif oper=='*':
    print(num1*num2)
elif oper=='/':
    if num1==0 or num2==0:
        print('두 수를 나눌 수 없습니다.')
    else:
        print(float(num1/num2))
```

```python
# 사칙연산 프로그램_2번째 풀이

num1=int(input('숫자1 : '))
num2=int(input('숫자2 : '))
oper=input('연산기호(+,-,*,/) : ')

if oper=='+':
    result = num1+num2
elif oper=='-':
    result = num1-num2
elif oper=='*':
    result = num1*num2
elif oper=='/':
    if num2==0:
        result='0 으로 나눌 수 없습니다.'
    else:
       result = float(num1/num2)

print(f'{num1} {oper} {num2} = {result}')
```
