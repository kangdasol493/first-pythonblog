```python
d_li = {}

while True:
    name = input('물건이름: ')
    if name == '':
        break
    num = int(input('num: '))
    if name in d_li.keys():
        d_li[name] += num
    else:
        d_li[name] = num
print(d_li)

for key,value in sorted(d_li.items()):
    print(key,value,end='')
```

```python
sweety = {'빼빼로', '칸초','허니버터'}
salty = {'포카칩', '새우깡','허니버터'}
#합집합
print(sweety.union(salty))
print(sweety|salty)
#교집합
print(sweety.intersection(salty))
print(sweety&salty)
#차집합
print(sweety.difference(salty))
print(sweety-salty)
print(salty.difference(sweety))
print(salty-sweety)
```

```python
# 할인율 계산
sale = ' '
price = int(input('물건 구매가를 입력하세요:'))
if price >= 100000:
    sale = '10'
elif price >= 50000:
    sale = '7.5'
elif price >= 10000:
    sale = '5'
else:
    sale = '0'
print(f'구매가: {price}원')
print(f'할인율: {sale}%')
print(f'할인금액: {int(price*sale/100)}원')
print(f'지불금액: {int(price-price*(sale/100))}원')
```

```python
num1 = int(input("숫자1: "))
num2 = int(input("숫자2: "))
mark = (input('연산기호(+,-,*,/): '))
result = 0

if mark == '+':
    result = num1 + num2
elif mark == '-':
    result = num1 - num2
elif mark == '*':
    result = num1 * num2
elif mark == '/':
    if num2 != 0:
        result = num1 / num2
    else:
        result = None

print('-'*30)
print(f'{num1}{mark}{num2}={result}')
```
